## 讀取資料

```python
import pandas
df = pandas.read_csv('data/20211109_top150_tw_stocks.csv')
df.set_index('Date', inplace=True)
df.head()
```

## 計算報酬

```python
df_stocks = pandas.pivot_table(df, index=df.index, values = 'Close', columns= 'symbol')    
df_change = df_stocks.pct_change()
```

## 繪製大盤走勢

```python
(df_change['^TWII'] + 1).cumprod().plot(kind = 'line', figsize = [20,10])
```

## 比較大盤與 2330 

```python
(df_change[['^TWII', '2330']] + 1).cumprod().plot(kind = 'line', figsize = [20,10])
```
